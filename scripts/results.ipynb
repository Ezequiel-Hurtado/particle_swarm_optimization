## 6. Under N dimensions

```python
for i in (2,3,5,10,20,50,100): 
    start_time = time.time()
    _,gb_img,_ = pso.pso_2d_dynamic(fct.sphere , N =i,s = 100, max_iter = 500, ub = 100, graph = 'Non')
    end_time = time.time()
    print(f"Dimensions : {i}")
    print(f"Image Global Best : {gb_img}")
    print(f"Time taken : {end_time-start_time}")
```
